```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
df = pd.read_csv('/home/trupti/Desktop/Zense_Proj_ML/archive/Data/features_3_sec.csv')
```

```python
print("Shape: ", df.shape)
print("")
print("First 5 audio files with features: ")
print("")
print(df.head())
```

```python
df.drop(labels='filename', axis=1, inplace=True)
```

```python
class_list = df.iloc[:, -1]
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(class_list)
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(np.array(df.iloc[:, :-1], dtype=float))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model = Sequential([
    Dense(512, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(256, activation='relu'),
    Dropout(0.2),
    Dense(128, activation='relu'),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(10, activation='softmax')
])
```

```python
def train_model_with_metrics(model, X_train, y_train, X_val, y_val, optimizer, loss_fn, metrics, epochs, batch_size):
    model.compile(optimizer=optimizer, loss=loss_fn, metrics=metrics)
    history = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=epochs, batch_size=batch_size, verbose=1)
    return history
```

```python
loss_function = 'sparse_categorical_crossentropy'
evaluation_metrics = ['accuracy']
batch_size = 128
model_history = train_model_with_metrics(model, X_train, y_train, X_test, y_test, 'adam', loss_function, evaluation_metrics, epochs=600, batch_size=batch_size)
print("Model Trained!")
```

```python
y_train_pred = model.predict(X_train).argmax(axis=1)
y_test_pred = model.predict(X_test).argmax(axis=1)
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
print("Training Accuracy for CNN model:", train_accuracy)
print("Testing Accuracy for CNN model:", test_accuracy)
```
